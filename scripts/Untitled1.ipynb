```python
# Celda 2: Importar y crear mapa
import leafmap. foliumap as leafmap

# Crear mapa centrado en Córdoba
m = leafmap.Map(center=[-31.4201, -64.1888], zoom=8)
m
```

```python
# Celda 3: Cargar departamentos de Córdoba
# El archivo está en /home/jovyan/data/cordoba/

m.add_geojson(
    '/home/jovyan/data/cordoba/departamentos.geojson',
    layer_name='Departamentos de Córdoba',
    style={'color': 'blue', 'weight': 2, 'fillOpacity':  0}
)
m
```

```python
# Celda 4: Cargar localidades desde CSV
import pandas as pd

# Leer CSV con localidades
localidades = pd.read_csv('/home/jovyan/data/cordoba/localidades.csv')

# Agregar puntos al mapa
m.add_points_from_xy(
    localidades,
    x='longitud',
    y='latitud',
    layer_name='Localidades',
    popup=['nombre', 'departamento']  # Columnas para el popup
)
m
```

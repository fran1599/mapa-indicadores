# 游댠 Mapas de Calor con Leafmap

Este notebook muestra c칩mo crear mapas de calor (heatmaps) para visualizar indicadores.

**Contenido:**
- Cargar datos de indicadores (con coordenadas incluidas)
- Crear mapa de calor b치sico
- Filtrar por tipo de indicador
- Guardar como HTML

```python
import leafmap.foliumap as leafmap
import pandas as pd
```

```python
# Cargar indicadores (ahora con coordenadas incluidas)
indicadores = pd.read_csv('/home/jovyan/data/ejemplo_indicadores.csv')
print(f"Registros: {len(indicadores)}")
print(f"Columnas: {indicadores.columns.tolist()}")
indicadores.head()
```

```python
# Crear mapa de calor
m = leafmap.Map(center=[-31.4201, -64.1888], zoom=7)
m.add_heatmap(
    data=indicadores,
    latitude='latitud',
    longitude='longitud',
    value='valor',
    name='Indicadores de Adicciones',
    radius=35
)
m
```

```python
# Filtrar por tipo
consumo = indicadores[indicadores['tipo_indicador'] == 'consumo']
m2 = leafmap.Map(center=[-31.4201, -64.1888], zoom=7)
m2.add_heatmap(data=consumo, latitude='latitud', longitude='longitud', value='valor', name='Consumo', radius=40)
m2
```

```python
# Guardar mapa
m.to_html('/home/jovyan/work/mapa_calor.html')
print("Mapa guardado en /home/jovyan/work/mapa_calor.html")
```
